```python
# Part 1 â€“ Mandelbrot and Julia Sets
import numpy as np
import matplotlib.pyplot as plt

def Mandelbrot(z,C):
    zNew = z**2 + C
    return zNew
```

```python
# Part 1 â€“ Mandelbrot and Julia Sets (1)
zmax = 30

n = 501 #Iteration number

x = np.linspace(-1.5,1.5,n)
y = np.linspace(-1.5,1.5,n)

C = [-0.4+0.6j]



def f(Z):
    for C1 in C:
        z_0 = Z
        for i in range(0,zmax):
            z_0 = Mandelbrot(z_0,C1)
            if abs(z_0) > 50:
                return 1
                break
        return 0
    return('did not diverge')
    
julia = np.zeros((n,n))


for a in range(n):
    for b in range(n):   
        Z = complex(x[a],y[b])
        julia[a,b]=f(Z)
        #if f(Z).z_0 >= zmax: #xlim ylim
            #print(a,b)
            #break
            
for i in range(len(x)):
    y[i] = f(x[i])
    
plt.plot(x,y,'-o')
plt.xlabel('Z values')
plt.ylabel('f(Z)')
plt.show()
    
    
plt.figure(figsize = (20,20))    
    
plt.xlim(0,500)
plt.ylim(0,500)
#plt.xlim(a,b)
#plt.ylim(a,b)

plt.imshow(julia)
```
