```python
# Here is a useful operator *=
a = 1
b = 
a *= 2
print(a)
print(b)
#the *= operator multiplies the left variable by the right operand and sets the left variable to this new multiplied value. In other words, (x *= 3) is (x = x * 3).
#four more two-symbol operators are +=, -=, /=, and %=.

c = 1
c += 3
print(c)
#this operator adds the left variable to the right operand. (c += 3) = (c = c+3).

d = 1
d -= 3
print(d)
#this operator subtracts the right operand from the left variable. (d -= 3) = (d = d-3).

e = 1
e /= 4
print(e)
#this operator divides the right operand from the left variable. (e /= 4) = (e = e/4).

f = 1
f %= 4
print(f)
#this operator takes the modulus of the right operand from the left variable. (f %= 4) = (f = f%4).
```

```python
# Let us apply the *= operator of a 1D Numpy array
import numpy as np
a = np.array([1, 2, 3, 3, 5, 5])
b = a
#a *= 2
#the behavior of the arrays is different from the behavior of the numbers above since two-symbol operators that have another variable referring to the variable at hand will recursively change all related variables accordingly. By using a single operator that refers to itself, only the variable in the command will be affected. As a result, this prevents the behavior of the numbers above from occurring.
a = a*2
print(a)
print(b)
```

```python
# 2D plots
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,361.0,1)
x2 = np.cos(np.radians(x))*3
y2 = np.cos(np.radians(2*x))*2
z2 = np.cos(np.radians(3*x))
plt.plot(x,x2*0,'k--',linewidth=1,dashes=(10,3));
plt.plot(x,x2,'m-.*',linewidth=3,markersize=0.2,mec='b',
         mew=0.1, mfc='blue',label="3cos(x)");
plt.plot(x,y2,'yD-',linewidth=2,markersize=1,mec='orange',
         mew=0.1, mfc='orange',label="2cos(2x)");
plt.plot(x,z2,'go--',linewidth=4,markersize=0.5,
         mec='g',mew=0.1, mfc='green',dashes=(5,1,1,1),label="cos(3x)");
plt.fill_between(x,y2,y2*0,color=(1.0,0.94,0.94),lw=0)
plt.xlabel('degrees (x-axis)')
plt.ylabel('value (y-axis)')
plt.legend()
plt.show()
```

```python
# More 2D plots
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 1.01, 0.01)
r = 0.5
x2 = x
y2 = r*x*(1-x)
r = 1.5
z2 = r*x*(1-x)
r = 2.5
a2 = r*x*(1-x)
r = 3.5
b2 = r*x*(1-x)
r = 4
c2 = r*x*(1-x)
plt.plot(x,x2,'m-.*',linewidth=3,markersize=0.2,mec='b',
         mew=0.1, mfc='blue',label="f(x)=x");
plt.plot(x,y2,'yD-',linewidth=2,markersize=1,mec='orange',
         mew=0.1, mfc='orange',label="g(x)=rx(1-x), r = 0.5");
plt.plot(x,z2,'go--',linewidth=4,markersize=0.5,
         mec='g',mew=0.1, mfc='green',dashes=(5,1,1,1),label="g(x)=rx(1-x), r = 1.5");
plt.plot(x,a2,'c:*',linewidth=3,markersize=0.2,mec='b',
         mew=0.1, mfc='blue',label="g(x)=rx(1-x), r = 2.5");
plt.plot(x,b2,'rD-',linewidth=2,markersize=1,mec='orange',
         mew=0.1, mfc='orange',label="g(x)=rx(1-x), r = 3.5");
plt.plot(x,c2,'ko--',linewidth=4,markersize=0.5,
         mec='g',mew=0.1, mfc='green',dashes=(5,1,1,1),label="g(x)=rx(1-x), r = 4");
plt.fill_between(x,y2,y2*0,color=(1.0,0.94,0.94),lw=0)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
#f(x) and g(x) touch at r = 0.5 and intersect at r = 1.5, r = 2.5, r = 3.5, and r = 4.
```

```python
# Even More 2D plots
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 1.01, 0.01)
r = 0.5
x2 = x
y2 = r*x*(1-x)
y3 = r*r*x*(1-x)*(1-r*x*(1-x))
r = 1.5
z2 = r*x*(1-x)
z3 = r*r*x*(1-x)*(1-r*x*(1-x))
r = 2.5
a2 = r*x*(1-x)
a3 = r*r*x*(1-x)*(1-r*x*(1-x))
r = 3.5
b2 = r*x*(1-x)
b3 = r*r*x*(1-x)*(1-r*x*(1-x))
r = 4
c2 = r*x*(1-x)
c3 = r*r*x*(1-x)*(1-r*x*(1-x))
plt.plot(x,x2,'m-.*',linewidth=3,markersize=0.2,mec='b',
         mew=0.1, mfc='blue',label="f(x)=x");
plt.plot(x,y3,'yD-',linewidth=2,markersize=1,mec='orange',
         mew=0.1, mfc='orange',label="h(x)=g(g(x))=rg(x)(1-g(x)), r = 0.5");
plt.plot(x,z3,'go--',linewidth=4,markersize=0.5,
         mec='g',mew=0.1, mfc='green',dashes=(5,1,1,1),label="h(x)=g(g(x))=rg(x)(1-g(x)), r = 1.5");
plt.plot(x,a3,'c:*',linewidth=3,markersize=0.2,mec='b',
         mew=0.1, mfc='blue',label="h(x)=g(g(x))=rg(x)(1-g(x)), r = 2.5");
plt.plot(x,b3,'rD-',linewidth=2,markersize=1,mec='orange',
         mew=0.1, mfc='orange',label="h(x)=g(g(x))=rg(x)(1-g(x)), r = 3.5");
plt.plot(x,c3,'ko--',linewidth=4,markersize=0.5,
         mec='g',mew=0.1, mfc='green',dashes=(5,1,1,1),label="h(x)=g(g(x))=rg(x)(1-g(x)), r = 4");
plt.fill_between(x,y2,y2*0,color=(1.0,0.94,0.94),lw=0)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
# Plotting the world’s CO2 record
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('ice_core_CO2_data.txt', usecols=(0,1)) 
plt.rcParams['figure.figsize'] = [15, 5]
plt.show() 
temp = data[:,1]
average = sum(temp) / temp.shape[0]
plt.rcParams['figure.figsize'] = [15, 5]
plt.plot(data[:,0],data[:,1],'r-',linewidth=1,
         markersize=0,mec='r',mew=2, mfc='pink',label="World's CO2 Record");
plt.fill_between(data[:,0],data[:,1],
                 average,color=(1.0,0.90,0.90),lw=0)
plt.xlabel('Time')
plt.ylabel('CO2 Contents')
plt.show()
```
