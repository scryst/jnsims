```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['animation.ffmpeg_path']=r'C:\Users\lhamr\OneDrive\Desktop\ffmpeg.exe'
%matplotlib inline
```

```python
# (PART 1)
# variables
cp = 1.17 # Joule/gramm/Kelvin
rho = 2.62 # gramm/cm^3
kappa = 0.031 # J/cm/seconds/Kelvin

k = kappa/(rho * cp)

print(f'{k} cm^2 s-1')
```

```python
# (PART 1) 
# interval = 50
intmult = 1
interval = 50*intmult
d = 500
width = d + 4*d

dx = d/interval

eta = k * 1200 / (dx*intmult)**2

dt = eta * dx**2 / k
print(dt)
print(f'eta= {eta}')

x = np.arange(0,width+dx,dx)
N = int(len(x))
# N0

T = np.zeros(N)

initial = int(2*interval)
final = int(3*interval) ## CHANGE ME

T[initial:final] = 1030

Tnew = np.copy(T)

print(T)
print(len(T))
plt.plot(x,T,'-o')
```

```python
iterations = int(3.154e7 / dt)
hour = int(iterations/8760)
day = int(iterations/365)
week = int(iterations/52)
month = int(iterations/12)
year = iterations

TTE = 0.0
for i in range(0,N):
    TTE += T[i]

for iteration in range(0,iterations+1):
    for i in range(0,N-1):
        Tnew[i]=eta*(T[i+1]-2*T[i]+T[i-1])+T[i]
    T=np.copy(Tnew)
    
    if iteration == year:
        midpoint = max(T)
        print(f'after 1 year, T= {midpoint}')
        LTE = 0
        for i in range(initial,final):
            LTE += T[i]
        print('\t fraction of lost thermal energy is', 1-LTE/TTE)
        
    if iteration == month:
        midpoint = max(T)
        print(f'after 1 month, T= {midpoint}')
        LTE = 0
        for i in range(initial,final):
            LTE += T[i]
        print('\t fraction of lost thermal energy is', 1-LTE/TTE)
    
    if iteration == week:
        midpoint = max(T)
        print(f'after 1 week, T= {midpoint}')
        LTE = 0
        for i in range(initial,final):
            LTE += T[i]
        print('\t fraction of lost thermal energy is', 1-LTE/TTE)
        
    if iteration == day:
        midpoint = max(T)
        print(f'after 1 day, T= {midpoint}')
        LTE = 0
        for i in range(initial,final):
            LTE += T[i]
        print('\t fraction of lost thermal energy is', 1-LTE/TTE)
        
    if iteration == hour:
        midpoint = max(T)
        print(f'after 1 hour, T= {midpoint}')
        LTE = 0
        for i in range(initial,final):
            LTE += T[i]
        print('\t fraction of lost thermal energy is', 1-LTE/TTE)
#after 1 hour, T= 1030.0
#	 fraction of lost thermal energy is 0.013889983412032292
#after 1 day, T= 1029.999994050111
#	 fraction of lost thermal energy is 0.06680842713501123
#after 1 week, T= 1005.0746560242926
#	 fraction of lost thermal energy is 0.1767063923175024
#after 1 month, T= 743.2790930951135
#	 fraction of lost thermal energy is 0.36299348465918524
#after 1 year, T= 248.85687596891725
#	 fraction of lost thermal energy is 0.762782407255457
```
